```python
import pandas as pd
```

```python
data = pd.read_csv('news_articles.csv')
data
```

```python
data.drop(['author', 'published', 'title', 'language', 'site_url', 'main_img_url', 'type', 'title_without_stopwords', 'text', 'hasImage'], axis=1, inplace=True)
```

```python
data
```

```python
data.info()
```

```python
## drop all the missing values in the dataset
fake_news_data= data.dropna()
```

```python
fake_news_data.info()
```

```python

# Import label encoder
from sklearn import preprocessing
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
# Encode labels in column 'Country'.
fake_news_data['label']= label_encoder.fit_transform(fake_news_data['label'])
print(fake_news_data.info())

# import numpy as np

# fake_news_data['label']=np.where(fake_news_data['label']== 'Fake', 0, 1)
```

```python
fake_news_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorize= CountVectorizer()

X= vectorize.fit_transform(fake_news_data.text_without_stopwords)
X= X.toarray()
X
```

```python
import numpy as np


y = np.array(fake_news_data.label)
y
```

```python
## split the test set
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=32)
```

```python
## apply Gaussian Naive Bayes model

from sklearn.naive_bayes import GaussianNB

NB_news= GaussianNB()
NB_model= NB_news.fit(X_train, y_train)
NB_model
```

```python
## make prediction

Y_pred_NB= NB_model.predict(X_test)
Y_pred_NB
```

```python
from sklearn.metrics import classification_report


print(classification_report(y_test, Y_pred_NB))
```

```python
## prepare confusion matrix



from sklearn.metrics import confusion_matrix


conf_matrix_NB= confusion_matrix(y_test, Y_pred_NB)
conf_matrix_NB
```

```python
## plot the confusion matrix

from matplotlib import pyplot as plt
from mlxtend.plotting import plot_confusion_matrix

fig, ax = plot_confusion_matrix(conf_mat=conf_matrix_NB, figsize=(4,4))
plt.xlabel('Predictions', fontsize = 18)
plt.ylabel('Actuals', fontsize = 18)
plt.title('Confusion Matrix', fontsize = 18)
plt.show()
```
